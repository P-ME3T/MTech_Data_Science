```python
def my_decorator(func):
    def wrapper():
        print('Before')
        func()
        print('After')
    return wrapper

def say_hi():
    print ('Hi')
```

```python
def my_decorator(func):
    def wrapper():
        print('Before')
        func()
        print('After')
    return wrapper

@my_decorator
def say_hi():
    print ('Hi')
    
say_hi()
```

```python
def do_twice(func):
    def wrapper_do_twice():
        func()
        func()
    return wrapper_do_twice

@do_twice
def say_hi():
    print ('Hi')
    
say_hi()
```

```python
#error
def do_twice(func):
    def wrapper_do_twice():
        func()
        func()
    return wrapper_do_twice

@do_twice
def greet(name):
    print(f"Hello {name}")
    
greet('John')
```

```python
import functools

def decorator(func):
    @functools.wraps(func)
    def wrapper_decorator(*args, **kwargs):
        value = func(*args, **kwargs)
        return value
    return wrapper_decorator
```

```python
import time

def timer(func):
    @functools.wraps(func)
    def wrapper_timer(*args, **kwargs):
        start_time = time.perf_counter()
        value = func(*args, **kwargs)
        end_time = time.perf_counter()
        run_time = end_time - start_time
        print(f"Finished {func.__name__!r} in {run_time: .4f} secs")
        return value
    return wrapper_timer

@timer
def waste(num_times):
    for _ in range(num_times):
        sum([i**2 for i in  range(10000)])
        
waste(1000)
waste.__doc__
timer.__doc__
```
